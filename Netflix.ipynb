##### The SQL table has already been created manually using a
##### CREATE TABLE statement with optimized data types and constraints.
##### Therefore, this script focuses only on inserting data and does
##### not create or modify the table schema.
##### The 'if_exists' parameter is intentionally set to 'append'
##### instead of 'replace' to ensure that:
- Existing table structure remains unchanged
- No table is dropped or recreated
- Data is safely appended into the empty structured table

```python
from sqlalchemy import NVARCHAR, VARCHAR, Integer
```

```python
df.to_sql(table_name,
          con=engine,
          if_exists='append',
          index=False,    
          dtype={
            "show_id": VARCHAR(10),
            "type": VARCHAR(10),
            "title": NVARCHAR(200),
            "director": VARCHAR(250),
            "cast": VARCHAR(1000),
            "country": VARCHAR(150),
            "date_added": VARCHAR(20),
            "release_year": Integer(),
            "rating": VARCHAR(10),
            "duration": VARCHAR(10),
            "listed_in": VARCHAR(100),
            "description": VARCHAR(500)
          }
)
```

# ---------------------------------------------------------
### Why SQLAlchemy Data Types Are Explicitly Imported
# ---------------------------------------------------------
##### The following SQLAlchemy data types are imported to explicitly
##### control how Pandas binds Python data to SQL Server columns
##### during the DataFrame.to_sql() operation.
#####
##### By default, Pandas may treat string columns as VARCHAR while
##### inserting data into SQL Server, even if the target SQL column
##### is defined as NVARCHAR.
#####
##### This behavior can cause loss of Unicode characters (e.g.,
##### Korean, Arabic, Chinese text), resulting in values being
##### stored as '???' in the database.
#####
##### To prevent this issue, SQLAlchemy data types are explicitly
##### provided to ensure correct Unicode handling and data integrity.
